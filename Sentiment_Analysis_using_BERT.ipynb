```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from bs4 import BeautifulSoup
import requests
import re
import torch
import numpy as np
import pandas as pd
```

```python
model_name="nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer= AutoTokenizer.from_pretrained(model_name)
model= AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
tokens=tokenizer.encode("I wish I were not here",return_tensors='pt')
```

```python
result=model(tokens)
```

```python
result
```

```python
int(torch.argmax(result.logits))+1
```

```python
r=requests.get(url="https://www.yelp.com/biz/tarragon-cafe-san-francisco")
```

```python
r
```

```python
soup=BeautifulSoup(r.text,"html.parser")
```

```python
soup
```

```python
regex=re.compile(".*comment.*")
results=soup.find_all('p',{'class':regex})
reviews=[result.text for result in results]
```

```python
results
```

```python
reviews
```

```python
df=pd.DataFrame(np.array(reviews),columns=['reviews'])
```

```python
df.tail()
```

```python
df.iloc[0]
```

```python
df['reviews'].iloc[0]
```

```python
def sentiment_score(review):
  tokens=tokenizer.encode(review,return_tensors='pt')
  result=model(tokens)
  return int(torch.argmax(result.logits))+1
```

```python
sentiment_score(df['reviews'].iloc[0])
```

```python
df['sentiment']=df['reviews'].apply(lambda x: sentiment_score(x[:512]))
```
